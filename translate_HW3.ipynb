```
import pandas as pd
from googletrans import Translator
import translators as ts
```

```
df = pd.read_csv('/work/HW3/HW3_dataset.csv')
df.columns = df.columns.str.strip() # remove eventual trailing spaces in column names
df.head()
```

```
# Get categorical columns list
numerical_cols = []
categorical_cols = []
for col in df.columns:
    if(df[col].dtype == 'object'):
        categorical_cols.append(col)
    else:
        numerical_cols.append(col)

print("List of categorical features : ", categorical_cols)
```

```
# Perform some basic cleaning
for col in categorical_cols:
    df[col] = df[col].str.lower() # to lowercase
    df[col] = df[col].str.strip() # remove trailing spaces
```

```
translator = Translator()
```

```
df['Occupation'] = df['Occupation'].astype(str)
result_occ = translator.translate(df['Occupation'].tolist(), src='tr', dest='en')
```

```
new_occ = []
for trans in result_occ:
    new_occ.append(trans.text)
df['Occupation_trans'] = new_occ
```

```
df['Education'] = df['Education'].astype(str)
result_edu = translator.translate(df['Education'].tolist(), src='tr', dest='en')
```

```
new_edu = []
for trans in result_edu:
    new_edu.append(trans.text)
df['Education_trans'] = new_edu
```

```
df['Working Status'] = df['Working Status'].astype(str)
result_work = translator.translate(df['Working Status'].tolist(), src='tr', dest='en')
```

```
new_work = []
for trans in result_work:
    new_work.append(trans.text)
df['Working_status_trans'] = new_work
```

```
df.to_csv('/work/HW3/HW3_dataset_translated.csv')
```
